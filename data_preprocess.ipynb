```python
import pandas as pd
import numpy as np
```

```python
car = ["audi", "bmw", "ford"]

data = []
for i in car:
    data.append(pd.read_csv("DataSet/"+ i +".csv"))
data = pd.concat(data, ignore_index=True)
```

```python
data
```

```python
# 移除異常值
data.drop(data[data["year"]>2022].index, inplace=True)
```

```python
# model處理
model_list = list(set(data['model']))
model_dict = dict()  
for i in range(0, len(model_list)):
    model_dict[model_list[i]]=i   
    
data["model"] = data["model"].map(model_dict)
```

```python
# year處理
year_list = list(set(data['year']))
year_dict = dict()  
now = max(year_list)
for i in range(0, len(year_list)):
    year_dict[year_list[i]]=now-year_list[i]  
    
data["year"] = data["year"].map(year_dict)
```

```python
# transmission 處理
transmission_dict = {'Manual': 0, 'Automatic': 1, "Semi-Auto":2}
data["transmission"] = data["transmission"].map(transmission_dict)
```

```python
# mileage 標準化
mileage = data['mileage']
mileage_mean = mileage.mean()
mileage_std = mileage.std()
data['mileage'] = (data['mileage'] - mileage_mean) / mileage_std
```

```python
# fuelType 處理
fuelType_list = list(set(data['fuelType']))

fuelType_dict = {'Petrol': 0, 'Diesel': 1, "Hybrid":2, "Other":3, "Electric":4}
data["fuelType"] = data["fuelType"].map(fuelType_dict)
```

```python
# tax 處理
tax = data['tax']
tax_mean = tax.mean()
tax_std = tax.std()
data['tax'] = (data['tax'] - tax_mean) / tax_std
```

```python
# mpg 處理
mpg = data['mpg']
mpg_mean = mpg.mean()
mpg_std = mpg.std()
data['mpg'] = (data['mpg'] - mpg_mean) / mpg_std
```

```python
# engineSize 處理
engineSize = data['engineSize']
engineSize_mean = engineSize.mean()
engineSize_std = engineSize.std()
data['engineSize'] = (data['engineSize'] - engineSize_mean) / engineSize_std
```

```python
data
```

```python
data.to_csv("DataSet/preprocessed data.csv",index=False)
```
