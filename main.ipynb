```python
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras import optimizers
from tensorflow.keras.models import  Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

data = pd.read_csv("DataSet/preprocessed data.csv")

y_train = data['price'].values
x_train = data.drop(['price'], axis=1).values
```

```python
adam = optimizers.Adam(learning_rate=0.001)

model = Sequential()
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

model.compile(optimizer = adam, loss='MAE', metrics=['mae'])
```

```python
early_stop = EarlyStopping(monitor='val_mae', patience=30, verbose=0, mode="auto")
reduce_lr = ReduceLROnPlateau(monitor='val_mae', factor=0.5, patience=20)
history = model.fit(x_train, y_train, epochs=2000, batch_size=128, verbose=2, validation_split=0.2, callbacks=[early_stop, reduce_lr], shuffle=True) 
```

```python
plt.plot(history.history['val_mae'], label='val_mae')
plt.plot(history.history['mae'], label='mae')
plt.title("MAE Trend")
plt.ylabel('MAE')
plt.xlabel('epoch')
plt.legend(loc = 0)
plt.show()
```
