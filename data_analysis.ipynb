```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
car = ["audi", "bmw", "ford"]
color = ["red", "green", "blue"]
```

```python
pd.read_csv("DataSet/audi.csv")
```

```python
#二手價格與里程數之關係
plt.figure(figsize=(8,5))

for i in range(0,3):
    data = pd.read_csv("DataSet/"+ car[i] +".csv")[["price","mileage"]]
    plt.scatter("price", "mileage", s=3, color = color[i], data=data, label = car[i], alpha = 1)

plt.title("Comparison of prices and mileage between different types of cars")
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.legend()
plt.show()
```

```python
#二手價格與註冊年份之關係
plt.figure(figsize=(8,5))

for i in range(0,3):
    data = pd.read_csv("DataSet/"+ car[i] +".csv")[["price", "year"]]
    data = data[data["year"]<2021]
    plt.scatter("price", "year", s=2, color = color[i], data=data, label = car[i], alpha = 1)

plt.title("Comparison of prices and years between different car types")
plt.xlabel('Price')
plt.ylabel('Year')
plt.legend()
plt.show()
```

```python
#二手價格與路稅比較
plt.figure(figsize=(8,5))

for i in range(0,3):
    data = pd.read_csv("DataSet/"+ car[i] +".csv")[["price", "tax"]]
    plt.scatter("price", "tax", s=2, color = color[i], data=data, label = car[i], alpha = 1)

plt.title("Comparison of prices and taxes between different types of cars")
plt.xlabel('Price')
plt.ylabel('Tax')
plt.legend()
plt.show()
```

```python
#二手價格與每加侖英里比較
plt.figure(figsize=(8,5))

for i in range(0,3):
    data = pd.read_csv("DataSet/"+ car[i] +".csv")[["price", "mpg"]]
    plt.scatter("price", "mpg", s=2, color = color[i], data=data, label = car[i], alpha = 1)

plt.title("Comparison of prices and Mpg between different types of cars")
plt.xlabel('Price')
plt.ylabel('Mpg')
plt.legend()
plt.show()
```
